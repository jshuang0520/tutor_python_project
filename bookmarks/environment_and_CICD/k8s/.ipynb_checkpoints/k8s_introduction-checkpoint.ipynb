# Introduction


<p style="text-align: left; font-family: 'objektiv-mk1,Helvetica,Arial,sans-serif,Linotte Light'; white-space:pre-wrap; word-wrap:break-word;"><font color='grey' size='4px'>Kubernetes（K8S）是一個可以幫助我們管理微服務（microservices）的系統，他可以自動化地部署及管理多台機器上的多個容器（Container）。簡單來說，他可以做到：
        
同時部署多個容器到多台機器上（Deployment）

服務的乘載量有變化時，可以對容器做自動擴展（Scaling）

管理多個容器的狀態，自動偵測並重啟故障的容器（Management）</font>
</p>

            </p>
            <p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;">
<font color='#52bd86' size='2px'>Kubernetes 運作的最小硬體單位，一個 Worker Node（簡稱 Node）對應到一台機器</font>
每個 Node 中都有三個組件：kubelet、kube-proxy、Container Runtime。
---
- <font color='#308df7'>kubelet</font>: 該 Node 的管理員，負責管理該 Node 上的所有 Pods 的狀態並負責與 Master 溝通
- <font color='#308df7'>kube-proxy</font>: 該 Node 的傳訊員，負責更新 Node 的 iptables，讓 Kubernetes 中不在該 Node 的其他物件可以得知該 Node 上所有 Pods 的最新狀態
- <font color='#308df7'>Container Runtime</font>: 該 Node 真正負責容器執行的程式，以 Docker 容器為例其對應的 Container Runtime 就是 Docker Engine
            </p>

一個 Master Node（簡稱 Master）中有四個組件：
kube-apiserver、etcd、kube-scheduler、kube-controller-manager。
---
- <font color='#308df7'>kube-apiserver</font>: <font color='#52bd86' size='2px'>負責 Node 之間的溝通橋樑，每個 Node 彼此不能直接溝通，必須要透過 apiserver 轉介</font>
- <font color='#308df7'>etcd</font>: 用來存放 Kubernetes Cluster 的資料作為備份，當 Master 因為某些原因而故障時，我們可以透過 etcd 幫我們還原 Kubernetes 的狀態
- <font color='#308df7'>kube-scheduler</font>: 負責監視 Cluster 狀態的 Process，例如：Node Controller、Replication Controller．這些 Process 會在 Cluster 與預期狀態（desire state）不符時嘗試更新現有狀態（current state）。來決定是否需要多開一台機器以應付突然增加的流量
- <font color='#308df7'>kube-controller-manager</font>: 整個 Kubernetes 的 Pods 調度員，scheduler 會監視新建立但還沒有被指定要跑在哪個 Node 上的 Pod，並根據每個 Node 上面資源規定、硬體限制等條件去協調出一個最適合放置的 Node 讓該 Pod 跑
            </p>

                <font color='grey' size='5px'>4. Cluster</font>
            </p>

Kubernetes 中多個 Node 與 Master 的集合。基本上可以想成在同一個環境裡所有 Node 集合在一起的單位。

#### API Version

<font color='#0091fa'>該 'Kubernetes元件' 的版本號</font>

This will determine the <font color='#52bd86'>version</font> of the <font color='#52bd86' size='3px'>Kubernetes Object</font>.

each resource will belong to a particular API group. For example, a Pod belongs to the API group /api/v1 , A deployment belongs to apps API group and so on...

#### Kind

<font color='#0091fa'>該元件是什麼屬性，常見有 Pod、Node、Service、Namespace、ReplicationController 等</font>

What <font color='#52bd86'>resource</font> are you creating? Pod? Service? Or maybe a Job? Specify within the "Kind" section.

You can create a lot of resources like pod, service, ingress, config map, daemon, job, cronjob, secret...the list goes on.

> 基本資源說明(KIND)
> - Namespace: 分隔各個團隊或專案資源的方法, 同類的資源名稱在同namespace內是唯一的
> - Deployment: 描述Deployment中的目標狀態, 定義你的app需要使用的資源
> -  <font color='#52bd86'>Pod: kubernetes中最小的編排單位</font>, 通常是一個或以上的容器組成, 這些容器通常共享同個network, file system
> - ConfigMap: 存放變數的資源
> - Secret: 存放機密資訊的資源
> -  <font color='#52bd86'>Service: 為一組pods提供相同的DNS, 類似於pods的load balancing</font>
> -  <font color='#52bd86'>Ingress: 將服務透出kubernetes的一種資源, 可以提供load balancing, SSL, virtual hosting</font>

#### Metadata

<p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;"><font color='#0091fa'>
- name: 指定該 Pod 的名稱
- labels: 指定該 Pod 的標籤，這裡我們暫時幫它上標籤為 app: demoApp
</font>
</p>

Look at this as some external data that is required so that you can <font color='#52bd86'>uniquely identify your resource</font>.  You can have a lot of pods running on your cluster but which one was created by you? Give your pod a name and you'll know.  Also maybe provide a namespace. <font color='#52bd86'>The name and the namespace are examples of metadata that you can easily provide!</font>


#### Spec

<p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;"><font color='#0091fa'>
- container.name: 指定運行出的 Container 的名稱
- container.image: 指定 Container 要使用哪個 Image，這裡會從 DockerHub 上搜尋
- container.ports: 指定該 Container 有哪些 port number 是允許外部資源存取
</font>
</p>

The spec <font color='#52bd86'>describes what state your resource should be in</font>. Like for example,

what image do you want to pull?

How many replicas do you want of your object?

Are there any specific ports, maybe environment variables or command arguments required?

## Kubernetes 進階三元件: Service、Deployment、Ingress

### Service

<p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;"><font color='#0091fa'>
但如果我們有多個 Pods 想要同時被連線時，我們就可以用到 Service 這個進階元件。
簡單來說，Service 就是 Kubernetes 中用來定義「一群 Pod 要如何被連線及存取」的元件
</font>
</p>

> 要建立一個 Service，一樣要撰寫屬於他的身分證 `service.yaml`

### Deployment

<p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;"><font color='#0091fa'>
把一個Pod做橫向擴展(複製多個相同的Pod在Cluster中同時提供服務)，
並監控如果有 Pod 當機我們就要重新把它啟動
</font>
</p>

<font color='grey'>今天當我們同時要把一個 Pod 做橫向擴展，也就是複製多個相同的 Pod 在 Cluster 中同時提供服務，並監控如果有 Pod 當機我們就要重新把它啟動時，如果我們要一個 Pod 一個 Pod 透過指令建立並監控是很花時間的。因此，我們可以透過 Deployment 這個特殊元件幫我們達成上述的要求。</font>



<font color='#0091fa' size='5px'>使用 Deployment 的好處:</font>

<font color='#0091fa' size='4.5px'>1. 自動部署應用</font>

一個應用一定會不斷的更新，Deployment 可以使用 Rolling Update(滾動式更新) 來不斷的更新 Pod 和部署應用。

<font color='#0091fa' size='4.5px'>2. Pod 更新過程不停機 (zero downtime)</font>

當 Pod 需要更新時，會先產生一個新的 Replica Set，並在新的 Replica Set 產生一個新的 Pod，接著會在舊的 Replica Set 刪除一個舊的 Pod ，重複這個動作直到所有需要更新的 Pod 都完成。因此更新的過程完全不會發生停機的問題，Pod 的數量也都維持不會少於指定的數量。

    <td>
        <p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;">Replica Set
Deployment 其實是搭配了 Replica Set 這項功能來使用的，Replica Set 是用來確保 Pod 的數量會一直和使用者指定的一樣。Replica Set 會在 Pod 故障時重新啟動新的 Pod，並把舊的 Pod 刪除。
而 Deployment 是更高層次的概念，他會管理建立出來的 Replica Set，如下圖 :</p>

<font color='#0091fa' size='3px'>apiVersion</font>
            
            該元件的版本號

<font color='#0091fa' size='3px'>kind</font>
            
            該元件是什麼屬性，
            常見有 Pod、Node、Service、Namespace、ReplicationController 等

<font color='#0091fa' size='3px'>metadata</font>
            
<font color='#00CC66' size='2px'> - name</font>
            
            指定該 Pod 的名稱

<font color='#0091fa' size='3px'>spec</font>
            
<font color='#00CC66' size='2px'> - replicas</font>
            
            指定要建立多少個相同的 Pod，
            在這邊給的數字是所謂的 Desire State，
            當 Cluster 運行時如果 Pod 數量低於此數字，Kubernetes 就會自動幫我們增加 pod，
            反之就會幫我們關掉 Pod


<font color='#00CC66' size='2px'> - template</font>
            
            解釋1:
            
            指定這個 Deployment 建立的 Pod 們統一的設定，
            包括 metadata 以及這些 Pod 的 Containers，
            這邊我們就沿用之前建立 Pod 的設定
            
            解釋2:
            
            template: 定義這個 Deployment 在建立 Pod 時的統一設定，
            Deployment 建立時就會依照這個樣板做出多個一模一樣的 Pod。

<font color='#00CC66' size='2px'> - selector</font>
            
            解釋1:
            
            指定這個 Deployment 的規則要適用到哪些 Pod，在這邊就是指定我們在 template 中指定的 labels
            
            解釋2:
            
            這裡的 Selector 定義是要讓 Deployment 知道他要管理哪些 Pod，也是用給標籤的概念。
</p>

### Ingress

> Service 就是 Kubernetes 中用來定義「一群 Pod 要如何被連線及存取」的元件

> Ingress 可以透過 HTTP/HTTPS，在我們眾多的 Service 前搭建一個 reverse-proxy。這樣 Ingress 可以幫助我們統一一個對外的 port number，並且根據 hostname 或是 pathname 決定封包要轉發到哪個 Service 上，如同下圖的比較：

<table>
    <td>
        <p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;">
            Ingress 用於將所有進入 Service 的 Port 統一，
            回想一下前面我們介紹 Service 時會制定開放的 Port，
            但是一旦 Service 的數量越來越多時就很難去維護了。

            只要對外開放的 Port 統一，

            如下圖 :</p>

<table>
    <td>
        <td width='300'>
        <p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;"><font color='grey' size='4px'>
apiVersion: networking.k8s.io/v1beta1
kind: Ingress
metadata:
  name: [ingress name]
spec:
  rules:
  - host: [domain name]
  http:
      paths:
      - path: /[path]
        backend:
          serviceName: [service name]
          servicePort: 80
  - https:
      paths:
      - path: /[path]
        backend:
          serviceName: [service name]
          servicePort: 443
            </font>
            </p>
        <p style="text-align: center;">a simple yaml config</p>
    </td>
    <td>
        <p style="vertical-align: text-top; text-align: left; font-family: 'objektiv-mk1'; white-space:pre-wrap; word-wrap:break-word;">
<font color='#00CC66'>paths :</font>

<font color='#00CC66'>- host</font>
            
            host 可以指定自訂的 Domain Name，
            也可以不設定就像上方 yaml 檔基本格式中 rules 的第二個沒有指定。
            沒有指定的情況就是使用主機的 IP。

<font color='#00CC66'>- path</font>
            
            path 可以定義特定的路徑會導向到指定的 Service，
            也可以不指定。

<font color='#00CC66'>- pathType</font>
            
            在 networking.k8s.io/v1beta1 這版是沒有支援 pathType，
            但是官方文件使用 networking.k8s.io/v1 這版有介紹，就順便紀錄下來。
            
            pathType 可以指定符合的路徑格式，
            例如，指定 Prefix 就代表是前面的 IP 會是主機的 IP，或是 Domain 會是指定的 Host，
            若沒有指定 pathType 則 Prefix 為預設值。
            也可以指定 Exact，就代表整個 URL 都要完全符合。pathType 更多用法可以參考 Kubernetes 官網。

<font color='#00CC66'>- backend</font>
            
            backend 可以指定後續要進行處理的 Service。
        
</p>
