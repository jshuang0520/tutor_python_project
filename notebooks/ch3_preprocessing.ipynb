```python
# check na ratio
def na_check(df_data):
    data_na = df_data.isnull().sum()
    # print('number of nan:\n{}'.format(list(zip(data_na.index, data_na))))
    print('number of nan:\n{}'.format(data_na))
    data_na = (data_na / len(df_data)) * 100
    data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending=False)
    missing_data = pd.DataFrame({'missing_ratio': data_na}).reset_index().rename(columns={'index': 'column_name'})
    # display(missing_data.head(10))
    return missing_data
```

```python
df_na_table = na_check(df_train)
df_na_table
```

```python
def check_na_rows(df, specific_col=None):
    if specific_col:
        df_res = df.loc[df[specific_col].isnull()]
    else:
        df_res = df.loc[df.isnull().any(axis=1)]
    print('shape: {}'.format(df_res.shape))
    return df_res
```

```python
X = df_train.loc[:, ['SibSp', 'Fare']].values  # .values -> map the data into a vector space (vectorize)
y = df_train.loc[:, ['Survived']].values

# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=87)

print('type(X_train):', type(X_train), '-', 'X_train[0:5]:', X_train[0:5], '-', 'X_train.shape:', X_train.shape)
```

```python
# feature scaling - method 1
from sklearn.preprocessing import StandardScaler
"""
這個方法是將資料值減掉平均值再除以標準差，標準化後，變數下的資料平均值等於0，標準差等於1(變異數等於1)。

這個方法保存了原始資料的分布型態，如果有異常值，異常值也會被保留。
適合用在常態分布的資料，對於不是常態分布資料，這不是最好的工具。
"""
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)

print('type(X_train):', type(X_train), '\n\n', 'X_train[0:5]:', X_train[0:5], '\n\n', 'X_train.shape:', X_train.shape)
```

```python
# feature scaling - method 2
from sklearn.preprocessing import MinMaxScaler
"""
它是將資料值減掉最小值再除以資料間距，資料間距是資料最大值減掉最小值，所以變數值收放至0到1，

這個方法可能不會維持原始資料的分布型態，且會受異常值影響。適用在標準差較小和不是常態分布資料。
"""
sc_X = MinMaxScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)

print('type(X_train):', type(X_train), '-', 'X_train[0:5]:', X_train[0:5], '-', 'X_train.shape:', X_train.shape)
```

- note.
  - sequential backward selection (SBS)
  
  > 循序地從特徵空間當中移除特徵，直到新的特徵空間只包含所想要的特徵個數．
  我們需定義一個最小化的「準則函數 (J)」(criterion function)，而準則定義為：模型在移除特定特徵前後預測效能的差異．
  我們要移除「影響預測效能最小」的特徵．
  
  > steps:
  >  1. 以 k=d 初始化演算法，其中 d 是全部特徵空間 Xd 的維數
  >  2. 找到讓「準則」(criterion) 最大的特徵 x- ; x- = argmax J(Xk - x), x 屬於 Xk
  >  3. 從特徵空間當中移除該特徵 x- ; Xk = (Xk) - (x-),  k = k-1
  >  4. 如果 k 等於所需特徵的個數則停止，否則回到 step 2 再繼續
  
  > [Comparative study of techniques for large-scale feature selection* 
Author links open overlay panel F.J.Ferria P.Pudilb M.Hatefc J.Kittlerc](https://www.sciencedirect.com/science/article/abs/pii/B9780444818928500407)

```python
# 0. 到底要選幾個變數才好
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification
# Create the RFE object and compute a cross-validated score.
svc = SVC(kernel="linear")
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(2),
              scoring='accuracy')
rfecv.fit(X_train, y_train)
print("Optimal number of features : %d" % rfecv.n_features_)
print("RFECV判斷過後，選 %s 個 變數 讓模型表現最好" % rfecv.n_features_)
# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
# SBS
# method 1 - sklearn

from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris


X, y = load_iris(return_X_y=True)
knn = KNeighborsClassifier(n_neighbors=3)
sfs = SequentialFeatureSelector(knn, n_features_to_select=3)
sfs.fit(X, y)
SequentialFeatureSelector(estimator=KNeighborsClassifier(n_neighbors=3),
                          n_features_to_select=3)
sfs.get_support()
sfs.transform(X).shape
```

```python
# SBS
# method 2 - mlxtend
'''
!pip install mlxtend
'''
from mlxtend.feature_selection import SequentialFeatureSelector as SFS


sbs = SFS(knn, 
          k_features=3,  # set an expected features to keep
          forward=False, 
          floating=False, 
          scoring='accuracy',
          cv=4,
          n_jobs=-1)

sbs = sbs.fit(X, y)  # sbs.fit(X, y)  # no need to use .values for X, y

print('- Sequential Backward Selection (k=3): {}'.format(sbs.k_feature_idx_))
print('- CV Score: {}'.format(sbs.k_score_))
print('- information: {}'.format(sbs.subsets_))

# import pprint
# pp = pprint.PrettyPrinter(indent=2)
# print('- information: \n', )
# pp.pprint(sbs.subsets_)
```

## methods

- [PCA, SVD, and LDA](https://blogs.sas.com/content/subconsciousmusings/2020/12/09/machine-learning-algorithm-use/)

> - We generally do not want to feed a large number of features directly into a machine learning algorithm since some features may be irrelevant or the “intrinsic” dimensionality may be smaller than the number of features. Principal component analysis (PCA), singular value decomposition (SVD), and Latent Dirichlet allocation (LDA) all can be used to perform dimension reduction. 

> - PCA is an unsupervised clustering method that maps the original data space into a lower-dimensional space while preserving as much information as possible. <font color='red'>The PCA basically finds a subspace that most preserve the data variance, with the subspace defined by the dominant eigenvectors of the data’s covariance matrix.</font>

--

- [the above context says LDA is 'Latent Dirichlet Allocation'](https://link.springer.com/chapter/10.1007/978-1-4614-3223-4_5#:~:text=Topic%20modeling%2C%20including%20probabilistic%20latent,in%20a%20collection%20of%20documents.)
> Topic modeling, including probabilistic latent semantic indexing and latent Dirichlet allocation, is a form of dimension reduction that uses a probabilistic model to find the co-occurrence patterns of terms that correspond to semantic topics in a collection of documents.
- [there is another LDA called 'Linear Discriminant Analysis'](https://machinelearningmastery.com/linear-discriminant-analysis-for-dimensionality-reduction-in-python/)
> Linear Discriminant Analysis, or LDA for short, is a predictive modeling algorithm for multi-class classification. It can also be used as a dimensionality reduction technique, providing a projection of a training dataset that best separates the examples by their assigned class.

--

- [principal component analysis (PCA)](https://builtin.com/data-science/step-step-explanation-principal-component-analysis)

> <font color='red'>it's about eigenvalues and eigenvectors</font>

> the first principal component accounts for the largest possible variance in the data set. 
> the first principal component ? Yes, it’s approximately the line that matches the purple marks because it goes through the origin and it’s the line in which the projection of the points (red dots) is the most spread out. Or mathematically speaking, it’s the line that maximizes the variance (the average of the squared distances from the projected points (red dots) to the origin).

> HOW DO YOU DO PCA?
> 1. Standardize the range of continuous initial variables
> 2. Compute the covariance matrix to identify correlations
> 3. Compute the eigenvectors and eigenvalues of the covariance matrix to identify the principal components
> 4. Create a feature vector to decide which principal components to keep
> 5. Recast the data along the principal components axes
