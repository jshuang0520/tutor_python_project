<font color='00CC66' size='5px'>steps to add a (monuted) disk to a current hadoop datanode</font>
    
<div class="alert-danger">
    
1. switch to hadoop user (be careful with the permission issue if we switch to root)

2. 'jps' to check the status of services, and then we stop the service

3. shutdown (the whole machine)

4. add disks (physically)

5. turn on the machine

6. mkdir -p /hdfs/2/datanode, and mount the disk 'permanently'! 
    
    then to fetch the mounted disk to hdfs-site: /opt/hadoop/etc/hadoop/hdfs-site.xml

</div>   

<div class="alert-success">
    
-> to /opt/hadoop/etc/hadoop/hdfs-site.xml

< property >

    <name>dfs.datanode.data.dir</name>

    <value>file:///hdfs/1/datanode,file:///hdfs/2/datanode</value>

< /property >

</div>

# elaboration    
---
        
<font color='grey' size='5px'>1. add a physical disk into a physical machine</font>

<font color='00CC66' size='3px'>1-0. switch the user account to root/hadoop</font>

$ sudo su -

<font color='00CC66' size='3px'>1-1. check hadoop datanode status, and stop its service</font>

$ jps

- Jps DataNode NodeManager

$ sudo systemctl stop datanode.service

$ jps
    
<font color='00CC66' size='3px'>1-2. shutdown the machine</font>
    
$ shutdown now
    
<font color='00CC66' size='3px'>1-3. add a disk into the machine</font>
    
<font color='00CC66' size='3px'>1-4. turn on the machine</font>
    
<font color='00CC66' size='3px'>1-5. start the service</font>
        
$ cd /etc/systemd/system
        
$ sudo vim datanode.service
        
$ sudo systemctl start datanode.service
    
$ jps
        
$ yarn app -list
        
$ yarn node -list -all
        
$ hadoop fs -node -list -all
        
$ yarn node -status [YOUR_HOST]:[YOUR_PORT]
        
$ hdfs --daemon stop datanode
        
$ yarn --daemon stop nodemanager
    
<font color='00CC66' size='3px'>1. add a physical disk into a physical machine</font>
        
<font color='00CC66' size='3px'>2. mount disks to a physical machine 'permanently'</font>
        
$ df -T
  
$ lsblk      -> to check storage of your disk
        
$ lsblk -f   -> to check NAME, FSTYPE, UUID, MOUNTPOINT of your disk
     
$ vim /etc/fstab    
        
<div class="alert-success">

UUID=[YOUR_DISK_UUID] /hdfs/2 ext4 inode_readhead_blks=128,commit=30,data=writeback,noatime,nodiratime,nodev,nobarrier 0 0

</div>
