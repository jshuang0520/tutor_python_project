- pull image

$ docker pull bitnami/spark

(https://hub.docker.com/r/bitnami/spark)

- run docker

$ docker run -d --name spark -e SPARK_MODE=master bitnami/spark

- get into the shell

$ docker exec -it <mycontainer> bash

- activate pyspark

$ pyspark
